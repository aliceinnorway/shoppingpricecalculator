```python
userStoryOne = "The value of an apple is 500"

#how many items are there
#add up price items
#find discount rate
#find tax rate
#calculate total order 

catalog = {
  "apple" : 500,
  "banana" : 1000,
  "lemon": 80,
  "lime": 40,
}

#Testing User Story One
print("User Story One: The value of an apple is ", catalog["apple"])

#User Story Two
userStoryTwo = "The shopping cart contains items"

shoppingCart = [catalog["apple"], catalog["apple"], catalog["apple"], catalog["banana"], catalog["banana"]]

shoppingCartValue = 0
for item in shoppingCart:
    shoppingCartValue += item

# Testing shopping cart value

testShoppingCart = 500 + 500 + 500 + 1000 + 1000
if testShoppingCart == shoppingCartValue :
    print ("User Story Two: Success ðŸŽ‰ The value of the shipping cart is ", shoppingCartValue)
elif testShoppingCart != shoppingCartValue :
    print ("User Story Two: Error in shopping value calculation")

    
# User Story Three - Calculate Discount Rate

userStoryThree = "Find the discount rate for the shopping cart value"

discountRate = 0

#this really should be in a def ...

if shoppingCartValue < 1000 :
    discountRate = 0
elif shoppingCartValue < 5000 :
    discountRate = 0.97
elif shoppingCartValue < 7000 :
    discountRate = 0.95
elif shoppingCartValue < 10000 :
    discountRate = 0.93
elif shoppingCartValue < 50000 :
    discountRate = 0.9
elif shoppingCartValue >= 50000 :
    discountRate = 0.85

#Testing user story three
print("User Story Three: The discount rate is :", discountRate)
    
# User Story Four
userStoryFour = "calculate tax rate"

taxRates = {
  "UT" : 0.0685,
  "NV" : 0.08,
  "TX" : 0.0625,
  "AL" : 0.04,
  "CA" : 0.0825,
}

# Insert mapping user location to tax rate

userLocation = "Texas"

if userLocation == "Utah":
    applicableTaxRate = taxRates["UT"]
    print ("User Story Four: The user is in Utah. The tax rate is", applicableTaxRate)
elif userLocation == "Nevada":
    applicableTaxRate = taxRates["NV"]
    print ("User Story Four: The user is in Nevada. The tax rate is", applicableTaxRate)
elif userLocation == "Texas":
    applicableTaxRate = taxRates["TX"]
    print ("User Story Four: The user is in Texas. The tax rate is", applicableTaxRate)
elif userLocation == "Alabama":
    applicableTaxRate = taxRates["AL"]
    print ("User Story Four: The user is in Alabama. The tax rate is", applicableTaxRate)
elif userLocation == "California":
    applicableTaxRate = taxRates["CA"]
    print ("User Story Four: The user is in California. The tax rate is", applicableTaxRate)

# User Story Five: add the discount

valueWithDiscount = shoppingCartValue * discountRate
print("User Story Five: The value with the discount" , valueWithDiscount)


# User Story Six: calculate Tax ðŸ’°

addedTax = applicableTaxRate * valueWithDiscount
print("User Story Six: The tax to be added it" , addedTax)  
#this needs a text here


# user Story seven: total order price

totalOrderValue = addedTax + valueWithDiscount
print("User Story Seven: The total order value is" , totalOrderValue) 
```
